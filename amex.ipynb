```python
def convert_to_float_and_replace_missing(data, col):
    for i in range(0, len(data)):
        if data[col].iloc[i]=='missing':
            data[col].iloc[i] = -1
    data[col] = data[col].astype(float)
    sum=0
    count=0
    for i in range(0, len(data)):
        if data[col].iloc[i]!=-1:
            sum = sum + data[col].iloc[i]
            count+=1
    sum = round(sum/count, 3)
    
    for i in range(0, len(data)):
        if data[col].iloc[i]==-1:
            data[col].iloc[i] = sum
```
